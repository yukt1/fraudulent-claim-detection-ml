### **6.1 Perform resampling** <font color="red">[3 Marks]</font>

Handle class imbalance in the training data by applying resampling technique.

Use the **RandomOverSampler** technique to balance the data and handle class imbalance. This method increases the number of samples in the minority class by randomly duplicating them, creating synthetic data points with similar characteristics. This helps prevent the model from being biased toward the majority class and improves its ability to predict the minority class more accurately.

**Note:** You can try other resampling techniques to handle class imbalance

```python
# Import RandomOverSampler from imblearn library
from imblearn.over_sampling import RandomOverSampler

# Perform resampling on training data
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)

# Verifying if the class imbalance is handled
print("Before resampling:", y_train.value_counts())
print("After resampling:", y_resampled.value_counts())
```

### **6.2 Feature Creation** <font color="red">[4 marks]</font>

Create new features from existing ones to enhance the model's ability to capture patterns in the data. This may involve deriving features from date/time columns, combining features, or creating interaction terms.

```python
# Create new features based on your understanding for both training and validation data
def feature_engineering(df):
    # Date features (year, month, weekday, weekend flag):
    # It is created as Fraud can follow time-based patterns (e.g., weekends, month-end spikes).
    df['incident_date'] = pd.to_datetime(df['incident_date'], errors='coerce')
    df['incident_year'] = df['incident_date'].dt.year
    df['incident_month'] = df['incident_date'].dt.month
    df['incident_day'] = df['incident_date'].dt.day
    df['incident_weekday'] = df['incident_date'].dt.weekday
    df['incident_is_weekend'] = df['incident_weekday'].isin([5, 6]).astype(int)

    # Claim ratio features (claim per customer month, injury/property/vehicle ratios):
    # It is created to Normalise raw claim amounts to reveal unusual proportions or high claims for new customers.
    df['claim_amount_per_month_customer'] = df['total_claim_amount'] / df['months_as_customer'].replace(0, 1)
    df['injury_ratio'] = df['injury_claim'] / (df['total_claim_amount'] + 1e-6)
    df['property_ratio'] = df['property_claim'] / (df['total_claim_amount'] + 1e-6)
    df['vehicle_ratio'] = df['vehicle_claim'] / (df['total_claim_amount'] + 1e-6)

    # Binary indicator features (property damage, umbrella policy, capital gain/loss flags):
    # It is created to Convert yes/no conditions into numeric data type that models can easily use.
    df['has_property_damage'] = df['property_damage'].apply(lambda x: 1 if str(x).strip().upper() == 'YES' else 0)
    df['has_umbrella'] = (df['umbrella_limit'] > 0).astype(int)
    df['has_capital_gain'] = (df['capital-gains'] > 0).astype(int)
    df['has_capital_loss'] = (df['capital-loss'] < 0).astype(int)

    # High-claim flag:
    # It is created as Fraudulent claims are often unusually large (with the box plot data skewed for higher amounts in fraud cases)
    # flagging the top 10% helps models detect these cases.
    high_claim_threshold = df['total_claim_amount'].quantile(0.9)
    df['high_claim'] = (df['total_claim_amount'] > high_claim_threshold).astype(int)

    return df
```

```python
X_train_fe = feature_engineering(X_resampled.copy())
X_val_fe = feature_engineering(X_val.copy())
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information
def combine_rare_categories(df, column, rare_cats, new_label='Other'):
    df[column] = df[column].replace(rare_cats, new_label)
    return df

# Detect categorical columns automatically
categorical_cols = X_train_clean.select_dtypes(include=['object', 'category']).columns.tolist()

# Dictionary to store rare category mappings
rare_category_maps = {}

# Process each categorical column
for col in categorical_cols:
    freq = X_train_clean[col].value_counts(normalize=True)
    rare_cats = freq[freq < 0.01].index
    rare_category_maps[col] = rare_cats
    X_train_clean = combine_rare_categories(X_train_clean, col, rare_cats)

# Apply same mapping to validation set
for col, rare_cats in rare_category_maps.items():
    X_val_clean = combine_rare_categories(X_val_clean, col, rare_cats)

```

### **7.3 Find the Optimal Cutoff** <font color = red>[12 marks]</font>

#### **7.3.1** Plot ROC Curve  to visualise the trade-off between true positive rate and false positive rate across different classification thresholds <font color="red">[2 Marks]</font>

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, auc

# Define ROC function
def plot_roc(y_true, y_pred_prob):
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6,6))
    plt.plot(fpr, tpr, color="blue", label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0,1], [0,1], color="red", linestyle="--")
    plt.xlabel("False Positive Rate (1 - Specificity)")
    plt.ylabel("True Positive Rate (Sensitivity)")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.show()

    # Return thresholds so we can use them later
    return fpr, tpr, thresholds

```

```python
# Call the ROC function
fpr, tpr, thresholds = plot_roc(y_train_enc, y_pred_prob)
```

## **Evaluation and Conclusion**

Data Quality and Issues:

The dataset contained both numeric and categorical variables, but several issues were identified. Some features, like capital-loss, had no variance and were removed, while others showed high collinearity or redundancy. The dataset also suffered from class imbalance, making fraud detection difficult. Better data collection could have ensured more balanced classes and reduced redundant or uninformative features.

Logistic Regression Performance:

The logistic regression model performed reasonably well on the training set, showing balanced sensitivity, specificity, and F1-scores. However, on the validation set it failed to detect fraud cases (0% sensitivity), despite maintaining high specificity. This indicates underfitting and limited ability to generalize to unseen data.

Random Forest Performance:

The random forest achieved perfect accuracy on the training data, showing clear signs of overfitting. On the validation set, however, it performed better than logistic regression, with sensitivity around 20% and precision of ~44%, producing a modest F1-score. While not ideal, it was more capable of detecting fraud than logistic regression.

Feature Importance Comparison:

Logistic regression emphasized categorical predictors such as policy_state, insured_education, and incident_severity. In contrast, random forest highlighted numeric, claims-related features like claim amount per month, injury ratio, vehicle ratio, and incident severity.

Agreement: Both models identified incident severity as important.

Difference: Logistic regression leaned on socio-demographics, while random forest prioritized behavioral/claims patterns.

Story Told by the Models:

The two models tell different stories. Logistic regression suggests fraud is influenced by demographic and policy attributes, whereas random forest suggests it is better explained by claim behaviors and ratios. Both perspectives are valid, but random forest captures more complex interactions.

Optimal Model and Conditions:

Between the two, random forest is the better model, as it was able to detect fraudulent cases in validation data, unlike logistic regression. The optimal conditions for random forest came from hyperparameter tuning: n_estimators=300, min_samples_split=2, min_samples_leaf=1, and max_features=log2. However to control the overfitting, the hyperparameter tuning is required to modify the factors such as max_depth, min_samples_split to get a better performing model.

Improving the Models Further:

Performance can be improved by addressing class imbalance (e.g., SMOTE, undersampling, or cost-sensitive learning), applying regularization or pruning to reduce overfitting, and experimenting with boosting algorithms (XGBoost, Gradient Boosting, CatBoost). Feature engineering to generate meaningful ratios, time-series features, or aggregated claim behavior will also strengthen predictive performance.


Conclusion:

While both models provide useful insights, the random forest model is recommended for deployment because it captures complex fraud patterns and performs better at detecting fraudulent cases. However, it still needs improvements to reduce overfitting and boost recall. Combining random forest with resampling techniques and boosting methods would likely yield a more robust fraud detection system.
